```python
from itertools import islice
from datasets import load_dataset, Dataset, DatasetDict

SUBSET_SIZE = 10_000                      # rows you want
OUT_DIR     = "/shared/public/data/rlucas/OpenR1-Math-10k"

# 1) Stream the full repo (no full download)
ds_stream = load_dataset("open-r1/OpenR1-Math-220k",
                         split="train", streaming=True)

# 2) Grab the first N rows (islice works with streaming datasets)
subset_rows = list(islice(ds_stream, SUBSET_SIZE))
train_ds    = Dataset.from_list(subset_rows)     # <-- correct call

# 3) (Optional) put it in a DatasetDict so GRPO-style code
#    that expects split names still works
train_ddict = DatasetDict({"train": train_ds})

# 4) Save locally
train_ddict.save_to_disk(OUT_DIR)
print(f"Wrote subset to {OUT_DIR}")
```
